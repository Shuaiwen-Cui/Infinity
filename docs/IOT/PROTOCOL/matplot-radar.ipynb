```python
# import libs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
metrics = ['data rate', 'range', 'scalability', 'reliability', 'security', 'power usage', 'cost']

# define a 2D numpy array, 13 rows, 7 columns

data = np.array([
    [  20,  10,  20,  80,  90,   5,  35], # nfc
    [  25,  15,  25,  75,  85,  10,  40], # rfid
    [  35,  30,  40,  70,  75,  60,  50], # bluetooth
    [  80,  80,  70,  85,  80,  70,  55], # wifi
    [  45,  40,  60,  75,  70,  50,  60], # zigbee
    [  10,  90,  50,  80,  65,  30,  75], # sigfox
    [  50,  50,  55,  75,  70,  40,  65], # thread
    [  15,  60,  65,  90,  80,  80,  80], # celluar_2g
    [  30,  70,  70,  85,  85,  75,  70], # celluar_3g
    [  70,  80,  75,  90,  90,  70,  65], # celluar_4g
    [ 100, 100,  80,  95,  95,  65,  60], # celluar_5g
    [  40,  65,  55,  80,  75,  20,  45], # nb_iot
    [  15,  85,  85,  75,  70,  15,  30]  # lora
])

# near field communication

nfc = data[0]
rfid = data[1]
bluetooth = data[2]

# cellular communication

cellular_2g = data[7]
cellular_3g = data[8]
cellular_4g = data[9]
cellular_5g = data[10]
nb_iot = data[11]

# non-cellular communication

wifi = data[3]
zigbee = data[4]
sigfox = data[5]
thread = data[6]
lora = data[12]
```

```python
angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False)
```

```python
# plot radar chart for the first cluster - near field communication

fig_near_field_communication, ax_near_field_communication = plt.subplots(nrows=1, ncols=1, subplot_kw=dict(polar=True))
ax_near_field_communication.set_theta_offset(np.pi/2)
ax_near_field_communication.set_theta_direction(-1)
ax_near_field_communication.set_thetagrids(np.degrees(angles), metrics)
ax_near_field_communication.set_ylim(0, 100)
ax_near_field_communication.set_title('Near Field Communication')
# use different colors for each protocol in the cluster
ax_near_field_communication.plot(angles, nfc, 'o-', linewidth=2, label='nfc', color='red')
ax_near_field_communication.plot(angles, rfid, 'o-', linewidth=2, label='rfid', color='green')
ax_near_field_communication.plot(angles, bluetooth, 'o-', linewidth=2, label='bluetooth', color='blue')
ax_near_field_communication.fill(angles, nfc, alpha=0.25, color='red')
ax_near_field_communication.fill(angles, rfid, alpha=0.25, color='green')
ax_near_field_communication.fill(angles, bluetooth, alpha=0.25, color='blue')
ax_near_field_communication.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
# fig_near_field_communication.savefig('near_field_communication.png', dpi=300, bbox_inches='tight')
# fig_near_field_communication.savefig('near_field_communication.jpg', dpi=300, bbox_inches='tight')

# show the plot
plt.show()
```
